```python
import pandas as pd
import json
```

```python
with open(r"E:\Users\Lucas xD\Downloads\Products_Q_US.json", encoding="utf8") as json_file:
    data = json.load(json_file)
```

```python
df = pd.json_normalize(data)
df.head()
```

```python
df = df.drop(columns=['careInstructions', 'styleNumber', 'categories'])
df.head()
```

```python
for index, product in df.iterrows():
        df.at[index,'image'] = product['variants'][0]['images'][0]
        product['longDescription'] = product['longDescription'].replace('<BR>', ' ')
        product['shortDescription'] = product['shortDescription'].replace('<BR>', ' ')
        product['longDescription'] = product['longDescription'].replace('<i>', ' ')
        product['shortDescription'] = product['shortDescription'].replace('<i>', ' ')
        df.at[index,'id'] = int(index)

        
```

```python
df.head(2)
```

```python
out = df.to_json(orient='records', indent=1)
with open(r"E:\Users\Lucas xD\Downloads\Products_Q_US_edited.json", 'w') as f:
    f.write(out)
```
